```python
import matplotlib.pyplot as plt

# Line plot for temperature anomalies
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Global_Temp_Anomaly'], label='Temperature Anomaly', color='blue', linewidth=1)
plt.title('Global Temperature Anomalies (1880 - 2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature Anomaly (°C)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate a rolling average (10-year window)
data['Rolling_Avg'] = data['Global_Temp_Anomaly'].rolling(window=10).mean()

# Line plot with rolling average
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Global_Temp_Anomaly'], alpha=0.5, label='Original Data')
plt.plot(data['Year'], data['Rolling_Avg'], color='red', linewidth=2, label='10-Year Rolling Average')
plt.title('Global Temperature Anomalies with Rolling Average', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature Anomaly (°C)', fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
# Filter data for the last 50 years
recent_data = data[data['Year'] >= 1975]

# Bar chart for recent years
plt.figure(figsize=(12, 6))
plt.bar(recent_data['Year'], recent_data['Global_Temp_Anomaly'], color='orange', alpha=0.7)
plt.title('Temperature Anomalies Over the Last 50 Years (1975 - 2024)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature Anomaly (°C)', fontsize=12)
plt.grid(axis='y')
plt.show()
```

```python
plt.annotate('Paris Agreement', xy=(2015, 0.8), xytext=(2005, 0.6),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
```

```python
# Pre-1950 warming rate
pre_1950 = data[data['Year'] < 1950]
pre_1950_rate = (pre_1950['Global_Temp_Anomaly'].iloc[-1] - pre_1950['Global_Temp_Anomaly'].iloc[0]) / len(pre_1950)

# Post-1950 warming rate
post_1950 = data[data['Year'] >= 1950]
post_1950_rate = (post_1950['Global_Temp_Anomaly'].iloc[-1] - post_1950['Global_Temp_Anomaly'].iloc[0]) / len(post_1950)

print("Pre-1950 warming rate:", pre_1950_rate)
print("Post-1950 warming rate:", post_1950_rate)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data for regression
X = data['Year'].values.reshape(-1, 1)
y = data['Global_Temp_Anomaly'].values

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict future anomalies
future_years = np.array([2025, 2030, 2040, 2050]).reshape(-1, 1)
predictions = model.predict(future_years)

print("Future Predictions:", predictions)
```
